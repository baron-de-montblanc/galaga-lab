```python
import numpy as np 
import sys
sys.path.append("../Backend")
from astro_objects import Galaxy
```

```python
gal_ellip = Galaxy(150.10, 2, z=0.15, mass=5e11, sed="elliptical", q=0.80)
gal_burst = Galaxy(150.10, 2, z=0.60, mass=8e10, sed="starburst", q=0.40)
gal_spiral = Galaxy(150.10, 2, z=0.35, mass=2e11, sed="spiral", q=0.55, agn_lum=3e10)

galaxies = np.array([gal_ellip, gal_burst, gal_spiral])
```

```python
for g, angle in zip(galaxies, (20, -35, 60)):
    g.angle = angle
```

```python
for g in galaxies:
    g.visualize()
```
